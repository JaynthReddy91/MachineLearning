```python
# Import libraries
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report
import matplotlib.pyplot as plt

# Load dataset
iris = load_iris()
X = iris.data
y = iris.target

# Split into train-test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# ---------------------------
# 1. Single Classifier - Decision Tree
# ---------------------------
dt = DecisionTreeClassifier(random_state=42)
dt.fit(X_train, y_train)
y_pred_dt = dt.predict(X_test)

# Evaluate Decision Tree
print("=== Decision Tree Performance ===")
print("Accuracy:", accuracy_score(y_test, y_pred_dt))
print("Precision:", precision_score(y_test, y_pred_dt, average='weighted'))
print("Recall:", recall_score(y_test, y_pred_dt, average='weighted'))
print("F1-score:", f1_score(y_test, y_pred_dt, average='weighted'))
print()

# ---------------------------
# 2. Ensemble Classifier - Random Forest
# ---------------------------
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)

# Evaluate Random Forest
print("=== Random Forest Performance ===")
print("Accuracy:", accuracy_score(y_test, y_pred_rf))
print("Precision:", precision_score(y_test, y_pred_rf, average='weighted'))
print("Recall:", recall_score(y_test, y_pred_rf, average='weighted'))
print("F1-score:", f1_score(y_test, y_pred_rf, average='weighted'))
print()

# ---------------------------
# 3. Effect of Number of Estimators
# ---------------------------
estimators = [1, 5, 10, 20, 50, 100, 200]
accuracies = []

for n in estimators:
    rf = RandomForestClassifier(n_estimators=n, random_state=42)
    rf.fit(X_train, y_train)
    y_pred = rf.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    accuracies.append(acc)

# Plot accuracy vs number of estimators
plt.figure(figsize=(8, 5))
plt.plot(estimators, accuracies, marker='o', color='b')
plt.title("Effect of Number of Estimators on Random Forest Accuracy")
plt.xlabel("Number of Estimators (Decision Trees)")
plt.ylabel("Accuracy")
plt.grid(True)
plt.show()

# ---------------------------
# Observations
# ---------------------------
print("=== Observations ===")
print("""
1. Random Forest generally outperforms a single Decision Tree across all metrics.
2. Increasing the number of estimators in Random Forest improves performance up to a certain point,
   after which the improvement becomes marginal.
3. Ensemble methods like Random Forest reduce overfitting by averaging results from multiple trees,
   leading to better generalization on unseen data.
4. Decision Trees can easily overfit, especially on small datasets, while Random Forest stabilizes
   predictions by leveraging diversity among trees.
""")

```

```python
# Import required libraries
import numpy as np
import pandas as pd
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import matplotlib.pyplot as plt

# ----------------------------
# Load Dataset
# ----------------------------
data = load_wine()
X = data.data
y = data.target

# Split data
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

# ----------------------------
# Function to evaluate model
# ----------------------------
def evaluate_model(model, X_test, y_test):
    y_pred = model.predict(X_test)
    return {
        "Accuracy": accuracy_score(y_test, y_pred),
        "Precision": precision_score(y_test, y_pred, average='weighted'),
        "Recall": recall_score(y_test, y_pred, average='weighted'),
        "F1-score": f1_score(y_test, y_pred, average='weighted')
    }

# ----------------------------
# 1. Effect of Number of Estimators
# ----------------------------
n_estimators_list = [5, 10, 50, 100, 200]
results_estimators = []

for n in n_estimators_list:
    rf = RandomForestClassifier(n_estimators=n, random_state=42)
    rf.fit(X_train, y_train)
    metrics = evaluate_model(rf, X_test, y_test)
    results_estimators.append(metrics["Accuracy"])

plt.figure(figsize=(7, 4))
plt.plot(n_estimators_list, results_estimators, marker='o', color='green')
plt.title("Effect of Number of Estimators on Accuracy")
plt.xlabel("Number of Estimators")
plt.ylabel("Accuracy")
plt.grid(True)
plt.show()

# ----------------------------
# 2. Effect of Maximum Depth
# ----------------------------
max_depth_list = [2, 4, 6, 8, None]
results_depth = []

for d in max_depth_list:
    rf = RandomForestClassifier(n_estimators=100, max_depth=d, random_state=42)
    rf.fit(X_train, y_train)
    metrics = evaluate_model(rf, X_test, y_test)
    results_depth.append(metrics["Accuracy"])

plt.figure(figsize=(7, 4))
plt.plot([str(d) for d in max_depth_list], results_depth, marker='o', color='blue')
plt.title("Effect of Maximum Depth on Accuracy")
plt.xlabel("Maximum Depth")
plt.ylabel("Accuracy")
plt.grid(True)
plt.show()

# ----------------------------
# 3. Effect of Random State (stability)
# ----------------------------
random_states = [0, 21, 42, 99, 123]
results_random = []

for r in random_states:
    rf = RandomForestClassifier(n_estimators=100, max_depth=6, random_state=r)
    rf.fit(X_train, y_train)
    metrics = evaluate_model(rf, X_test, y_test)
    results_random.append(metrics["Accuracy"])

plt.figure(figsize=(7, 4))
plt.plot(random_states, results_random, marker='o', color='red')
plt.title("Effect of Random State on Accuracy")
plt.xlabel("Random State")
plt.ylabel("Accuracy")
plt.grid(True)
plt.show()

# ----------------------------
# Final Model Evaluation
# ----------------------------
best_rf = RandomForestClassifier(n_estimators=100, max_depth=6, random_state=42)
best_rf.fit(X_train, y_train)
final_metrics = evaluate_model(best_rf, X_test, y_test)

print("=== Final Model Performance (n_estimators=100, max_depth=6) ===")
for k, v in final_metrics.items():
    print(f"{k}: {v:.4f}")

# ----------------------------
# Observations
# ----------------------------
print("\n=== Detailed Observations ===")
print("""
1. Number of Estimators:
   - Increasing the number of estimators (trees) improves model stability and accuracy.
   - After around 100 estimators, performance gains become marginal.
   - This demonstrates the bagging principle: averaging predictions across many trees reduces variance.

2. Maximum Depth:
   - Shallow trees (low depth) underfit the data and result in lower accuracy.
   - Deeper trees capture more complex patterns but may overfit slightly.
   - A moderate depth (around 6–8) provides a good trade-off between bias and variance.

3. Random State:
   - Changing the random seed changes the random sampling of data and feature subsets.
   - Accuracy remains quite stable, showing that bagging improves robustness to randomness.

4. Justification:
   - Bagging (Bootstrap Aggregating) reduces overfitting by training each tree on different random samples.
   - Combining multiple weak learners (Decision Trees) creates a strong and stable classifier.
   - RandomForest’s ensemble averaging smooths out noise and improves generalization.

Overall, the Bagging technique through RandomForestClassifier provides high accuracy,
low variance, and robustness against random fluctuations in data.
""")
```

```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier
from sklearn.metrics import accuracy_score
from sklearn.impute import SimpleImputer

# ----------------------------
# 1. Generate Synthetic Dataset (2 Features for Visualization)
# ----------------------------
X, y = make_classification(n_samples=400, n_features=2, n_redundant=0,
                           n_informative=2, n_clusters_per_class=1, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# ----------------------------
# 2. Define Ensemble Models
# ----------------------------
models = {
    "Random Forest": RandomForestClassifier(n_estimators=100, random_state=42),
    "AdaBoost": AdaBoostClassifier(n_estimators=100, random_state=42),
    "Gradient Boosting": GradientBoostingClassifier(n_estimators=100, random_state=42)
}

# ----------------------------
# 3. Train Models and Plot Decision Boundaries
# ----------------------------
fig, axes = plt.subplots(1, 3, figsize=(15, 4))
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02),
                     np.arange(y_min, y_max, 0.02))

for ax, (name, model) in zip(axes, models.items()):
    model.fit(X_train, y_train)
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    ax.contourf(xx, yy, Z, alpha=0.3, cmap=plt.cm.coolwarm)
    ax.scatter(X_train[:, 0], X_train[:, 1], c=y_train, edgecolor='k', cmap=plt.cm.coolwarm)
    ax.set_title(f"{name} Decision Boundary")

plt.suptitle("Decision Boundaries of Ensemble Models", fontsize=14)
plt.show()

# ----------------------------
# 4. Compare Performance Under Noisy Data
# ----------------------------
# Add Gaussian noise to test data
X_test_noisy = X_test + np.random.normal(0, 0.5, X_test.shape)

print("=== Model Performance with Noise ===")
for name, model in models.items():
    y_pred_noisy = model.predict(X_test_noisy)
    acc = accuracy_score(y_test, y_pred_noisy)
    print(f"{name}: Accuracy under noise = {acc:.4f}")
print()

# ----------------------------
# 5. Compare Performance with Missing Data
# ----------------------------
# Randomly insert NaN values
X_test_missing = X_test.copy()
mask = np.random.rand(*X_test_missing.shape) < 0.1  # 10% missing
X_test_missing[mask] = np.nan

# Impute missing values (mean strategy)
imputer = SimpleImputer(strategy='mean')
X_test_imputed = imputer.fit_transform(X_test_missing)

print("=== Model Performance with Missing Data ===")
for name, model in models.items():
    y_pred_missing = model.predict(X_test_imputed)
    acc = accuracy_score(y_test, y_pred_missing)
    print(f"{name}: Accuracy with missing data = {acc:.4f}")
print()

# ----------------------------
# 6. Feature Importance Analysis
# ----------------------------
plt.figure(figsize=(7, 5))
for name, model in models.items():
    importances = model.feature_importances_
    plt.barh([f"Feature {i+1}" for i in range(len(importances))],
             importances, alpha=0.6, label=name)

plt.title("Feature Importance Across Ensemble Models")
plt.xlabel("Importance Score")
plt.legend()
plt.show()

# ----------------------------
# 7. Observations
# ----------------------------
print("=== Observations ===")
print("""
1. Decision Boundaries:
   - Random Forest creates more irregular, detailed boundaries due to bagging and randomness.
   - AdaBoost and Gradient Boosting generate smoother, more directional boundaries as they iteratively
     correct previous mistakes (boosting).

2. Under Noisy Data:
   - Random Forest tends to handle noise better because averaging over many trees reduces variance.
   - Boosting models (AdaBoost, Gradient Boosting) can be more sensitive to noise since they focus on
     hard-to-classify points, which may include noisy samples.

3. With Missing Data:
   - After imputing, all models perform slightly worse but remain robust.
   - Random Forest remains the most stable because its trees can handle small data imperfections.

4. Feature Importance:
   - All ensemble models identify similar key features, but boosting models often assign sharper
     importance distinctions.
   - Random Forest spreads importance more evenly, reflecting its averaging nature.

Conclusion:
Ensemble models provide flexible and powerful decision boundaries.
Random Forest offers robustness under noise and missing data, while boosting methods can provide
higher accuracy on clean, well-defined datasets.
""")
```
